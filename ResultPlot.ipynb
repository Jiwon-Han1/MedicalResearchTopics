```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
root_dir = '/home/jiwon/MedicalBioResearch/Prediction_Sex/'
os.chdir(root_dir)
print(os.getcwd())
```

```python
# Plotting the Results
fig, axes = plt.subplots(1, 3, figsize=(18,6))
axes[0].set_title('Training Loss')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Loss')
axes[0].set_xticks(np.linspace(0,1,5)*20)

axes[1].set_title('Training Accuracy')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('Accuracy')
axes[1].set_xticks(np.linspace(0,1,5)*20)

axes[2].set_title('Validation Accuracy')
axes[2].set_xlabel('Epoch')
axes[2].set_ylabel('Accuracy')
axes[2].set_xticks(np.linspace(0,1,5)*20)

for num_mod in range(1,6):
    epoch_loss_values = np.loadtxt(os.path.join(root_dir, f'Loss_train_{num_mod}.txt'))
    epoch_metric_values = np.loadtxt(os.path.join(root_dir, f'Acc_train_{num_mod}.txt'))
    metric_values = np.loadtxt(os.path.join(root_dir, f'Acc_val_{num_mod}.txt'))
    
    axes[0].plot([i + 1 for i in range(len(epoch_loss_values))], epoch_loss_values, label=f'Training Loss: {num_mod} modalities')
    axes[1].plot([i + 1 for i in range(len(epoch_metric_values))], epoch_metric_values, label=f'Training Acc: {num_mod} modalities')
    axes[2].plot([(i + 1) for i in range(len(metric_values))], metric_values, label=f'Validation Acc: {num_mod} modalities')
    
axes[0].legend()
axes[1].legend()
axes[2].legend()

plt.tight_layout()
plt.savefig(os.path.join(root_dir, "Performance_Comparison.png"), dpi=300)

```
